## Concepts

1. Quantization - parameters are stored using either 16 or 32 binary digits. But, we can reduce it down to fewer digits, like eight bits, which means that there's only 256 possible values, or even four bits, meaning at a lower precision - turning it into something which has discrete possible values. This is done to squeeze more into less memory and the calculations are way simpler. This degrades the performance accuracy though.

-> bitsandbytes is the name of hugging face's library which deals with quantization

-> tokenizer EOS token comes at the end of the sentence and tokenizer Pad token is used to add extra spacing if it needs to.

tokenizer.pad_token = tokenizer.eos_token

2. Model Object - topmost is embedding, then layers and then at the very bottom there is LLM head.

-> embedding [n_in , n_out] - dimensionality of a layer. It is sometimes called vector embeddings or encoder layer. It takes the tokens that come in as the input. And for each of these tokens, it turns it into something which is a bunch of numbers that best represent its understanding of what that token represents. It's basically translating a single word into a vector with 2048 dimensions.

-> rotary embedding - a particular way which makes sure that words in a sequence that the order in which they appear is sort of affected in some way that's taken into account.

-> fully connected layer - it outputs the probability of the next token.

3. Decoder in LLaMA's architecture consists of basically three layers - 
-> chuncks attention layer

-> MLP, that is a multi-layer perceptron

-> Layer Norm

4. Introducing non-linearity becomes absolutely crucial to making sure that all the parameters matter.

-> The one function used in LLaMA is SiLU - sigmoid linear function.

5. Attention Mask - tells the model which token should this token pay attention to, and which one should be ignored.

6. Streaming - the model sends generated tokens one-by-one (or chunk-by-chunk) as they are produced, instead of waiting for the entire response to finish.
