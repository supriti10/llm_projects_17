````python
# ============================================================
# AI CODE GENERATION SUITE
# FEATURES:
# 1. Add Docstrings + Comments
# 2. Generate Unit Tests
# 3. Generate Trading Bots
#
# MODELS:
# - Groq -> llama-3.3-70b-versatile
# - Gemini -> gemini-2.5-flash-lite
#
# UI:
# - Gradio
# ============================================================

import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
````

````python
# ============================================================
# LOAD ENV
# ============================================================

load_dotenv()

groq_api_key = os.getenv("GROQ_API_KEY")
google_api_key = os.getenv("GOOGLE_API_KEY")
````

````python
# ============================================================
# CLIENTS
# ============================================================

groq = OpenAI(
    api_key=groq_api_key,
    base_url="https://api.groq.com/openai/v1"
)

gemini = OpenAI(
    api_key=google_api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

# ============================================================
# MODELS
# ============================================================

MODELS = {
    "Groq": "llama-3.3-70b-versatile",
    "Gemini": "gemini-2.5-flash-lite"
}

CLIENTS = {
    "Groq": groq,
    "Gemini": gemini
}
````

````python
# ============================================================
# SYSTEM PROMPTS
# ============================================================

DOCSTRING_PROMPT = """
You are an expert software engineer.

Your task:
- Add professional docstrings
- Add inline comments
- Improve readability
- Keep functionality EXACTLY the same

Return ONLY valid code.
"""

UNIT_TEST_PROMPT = """
You are a senior QA automation engineer.

Generate:
- pytest unit tests
- edge cases
- invalid inputs
- production-quality tests

Return ONLY valid Python pytest code.
"""

TRADING_PROMPT = """
You are an expert quantitative trading engineer.

Generate:
- production-ready trading bots
- risk management
- stop loss
- take profit
- clean architecture

Use a SIMULATED trading environment only.

Return ONLY valid Python code.
"""
````

````python
# ============================================================
# GENERATION FUNCTION
# ============================================================

def generate(task, model_name, user_input):

    client = CLIENTS[model_name]
    model = MODELS[model_name]

    # ========================================================
    # SELECT PROMPT
    # ========================================================

    if task == "Add Docstrings + Comments":
        system_prompt = DOCSTRING_PROMPT

    elif task == "Generate Unit Tests":
        system_prompt = UNIT_TEST_PROMPT

    else:
        system_prompt = TRADING_PROMPT

    # ========================================================
    # API CALL
    # ========================================================

    response = client.chat.completions.create(
        model=model,
        messages=[
            {
                "role": "system",
                "content": system_prompt
            },
            {
                "role": "user",
                "content": user_input
            }
        ]
    )

    result = response.choices[0].message.content

    # remove markdown
    result = (
        result
        .replace("```python", "")
        .replace("```", "")
        .strip()
    )

    return result
````

````python
# ============================================================
# SAMPLE INPUTS
# ============================================================

sample_code = """
def add(a,b):
    return a+b
"""

sample_strategy = """
Create a moving average crossover strategy.

BUY:
- 50 EMA crosses above 200 EMA

SELL:
- 50 EMA crosses below 200 EMA

Risk Management:
- stop loss 2%
- take profit 5%
"""
````

````python
# ============================================================
# UI
# ============================================================

with gr.Blocks(
    theme=gr.themes.Monochrome(),
    title="AI Code Generation Suite"
) as ui:

    gr.Markdown(
        "# AI Code Generation Suite"
    )

    # ========================================================
    # CONTROLS
    # ========================================================

    with gr.Row():

        task = gr.Dropdown(
            [
                "Add Docstrings + Comments",
                "Generate Unit Tests",
                "Generate Trading Bot"
            ],
            value="Add Docstrings + Comments",
            label="Task"
        )

        model = gr.Dropdown(
            ["Groq", "Gemini"],
            value="Groq",
            label="Model"
        )

    # ========================================================
    # INPUT
    # ========================================================

    user_input = gr.Code(
        label="Input",
        value=sample_code,
        language="python",
        lines=20
    )

    # ========================================================
    # BUTTONS
    # ========================================================

    with gr.Row():

        use_code_sample = gr.Button("Load Code Sample")

        use_strategy_sample = gr.Button("Load Strategy Sample")

        generate_btn = gr.Button("Generate")

    # ========================================================
    # OUTPUT
    # ========================================================

    output = gr.Code(
        label="Generated Output",
        language="python",
        lines=25
    )

    # ========================================================
    # SAMPLE LOADERS
    # ========================================================

    def load_code_sample():
        return sample_code

    def load_strategy_sample():
        return sample_strategy

    use_code_sample.click(
        fn=load_code_sample,
        outputs=[user_input]
    )

    use_strategy_sample.click(
        fn=load_strategy_sample,
        outputs=[user_input]
    )

    # ========================================================
    # GENERATE
    # ========================================================

    generate_btn.click(
        fn=generate,
        inputs=[task, model, user_input],
        outputs=[output]
    )
````

````python
# ============================================================
# LAUNCH
# ============================================================

ui.launch(inbrowser=True)
````
