## And now for some fun - an adversarial conversation between Chatbots..

You're already familar with prompts being organized into lists like:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "user prompt here"}
]
```

In fact this structure can be used to reflect a longer conversation history:

```
[
    {"role": "system", "content": "system message here"},
    {"role": "user", "content": "first user prompt here"},
    {"role": "assistant", "content": "the assistant's response"},
    {"role": "user", "content": "the new user prompt"},
]
```

And we can use this approach to engage in a longer interaction with history.

````python
# Let's make a conversation between Groq and Gemini
# We're using cheap versions of models so the costs will be minimal

import google.generativeai as genai
from groq import Groq
from IPython.display import Markdown, display

genai.configure(api_key='GOOGLE_API_KEY')
groq_client = Groq(api_key='GROQ_API_KEY')

groq_model = "llama-3.3-70b-versatile"
gemini_model = "gemini-2.5-flash-lite"

groq_system = "You are a chatbot who is very argumentative; \
you disagree with anything in the conversation and you challenge everything, in a snarky way."

gemini_system = "You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting."

groq_messages = ["Hi there"]
gemini_messages = ["Hi"]
````

````python
gemini_model_instance = genai.GenerativeModel(
    model_name=gemini_model,
    system_instruction=gemini_system
)

gemini_chat = gemini_model_instance.start_chat(history=[])
````

````python
def call_groq():
    messages = [{"role": "system", "content": groq_system}]
    for groq_msg, gemini_msg in zip(groq_messages, gemini_messages):
        messages.append({"role": "assistant", "content": groq_msg})
        messages.append({"role": "user", "content": gemini_msg})
    response = groq.chat.completions.create(model=groq_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_groq()
````

````python
def call_gemini():
    messages = [{"role": "system", "content": gemini_system}]
    for groq_msg, gemini_msg in zip(groq_messages, gemini_messages):
        messages.append({"role": "user", "content": groq_msg})
        messages.append({"role": "assistant", "content": gemini_msg})
    messages.append({"role": "user", "content": groq_messages[-1]})
    response = gemini.chat.completions.create(model=gemini_model, messages=messages)
    return response.choices[0].message.content
````

````python
call_gemini()
````

````python
call_groq()
````

````python
groq_messages = ["Hi there"]
gemini_messages = ["Hi"]

display(Markdown(f"### Groq:\n{groq_messages[0]}\n"))
display(Markdown(f"### Gemini:\n{gemini_messages[0]}\n"))

for i in range(5):
    groq_next = call_groq()
    display(Markdown(f"### Groq:\n{groq_next}\n"))
    groq_messages.append(groq_next)
    
    gemini_next = call_gemini()
    display(Markdown(f"### Gemini:\n{gemini_next}\n"))
    gemini_messages.append(gemini_next)
````

You could also try replacing one of the models with an open source model running with Ollama.

````python
#Imports

from groq import Groq
import google.generativeai as genai
from IPython.display import Markdown, display
````

````python
#Set api and constants

import os
from dotenv import load_dotenv

load_dotenv()

genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
groq_client = Groq(api_key=os.getenv("GROQ_API_KEY"))

GROQ_MODEL = "llama-3.3-70b-versatile"
GEMINI_MODEL = "gemini-2.5-flash-lite"
````

````python
#Define prompts

alex_system = """
You are Alex, a chatbot who is very argumentative; 
you disagree with anything in the conversation and challenge everything in a snarky way.
"""

blake_system = """
You are Blake, a very polite and agreeable chatbot.
You try to agree with others and keep things calm.
"""

charlie_system = """
You are Charlie, a neutral and analytical thinker.
You try to bring logic and balance into the discussion.
"""
````

````python
#Build prompt function

def build_user_prompt(name, others, conversation):
    return f"""
        You are {name}. 
        
        You are in a 3 person conversation with {others}.

        Conversation so far:
        {conversation}

        Now respond as {name}."""
````

````python
#Model calls

def call_groq(system_prompt, user_prompt):
    response=groq_client.chat.completions.create(
        model=GROQ_MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0.8
    )
    return response.choices[0].message.content


def call_gemini(system_prompt, user_prompt):
    model = genai.GenerativeModel(
        model_name=GEMINI_MODEL,
        system_instruction=system_prompt
    )
    response = model.generate_content(user_prompt)
    return response.text
````

````python
#Conversation state

conversation = ""

def add_message(speaker, message):
    global conversation
    conversation += f"{speaker}: {message}\n"
````

````python
#Initial message

add_message("Alex", "AI is making life easier? That's a lazy assumption.")
add_message("Blake", "Hey, I think it helps a lot actually.")
add_message("Charlie", "Well, that depends on it's context and implementation.")

display(Markdown("### Alex:\nAI is making life easier? That's a lazy assumption.\n"))
display(Markdown("### Blake:\nHey, I think it helps a lot actually.\n"))
display(Markdown("### Charlie:\nWell, that depends on it's context and implementation.\n"))
````

````python
#Loop

for _ in range(4):

    # Alex (talks to Blake & Charlie)
    alex_prompt = build_user_prompt("Alex", "Blake and Charlie", conversation)
    alex_reply = call_groq(alex_system, alex_prompt)
    add_message("Alex", alex_reply)
    display(Markdown(f"### Alex:\n{alex_reply}\n"))

    # Blake (talks to Alex & Charlie)
    blake_prompt = build_user_prompt("Blake", "Alex and Charlie", conversation)
    blake_reply = call_gemini(blake_system, blake_prompt)
    add_message("Blake", blake_reply)
    display(Markdown(f"### Blake:\n{blake_reply}\n"))

    # Charlie (talks to Alex & Blake)
    charlie_prompt = build_user_prompt("Charlie", "Alex and Blake", conversation)
    charlie_reply = call_groq(charlie_system, charlie_prompt)
    add_message("Charlie", charlie_reply)
    display(Markdown(f"### Charlie:\n{charlie_reply}\n"))
````
