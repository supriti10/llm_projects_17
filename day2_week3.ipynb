## The Two API Levels of Hugging Face

1. Pipelines - Higher level APIs to carry out standard tasks incredibly quickly

2. Tokenizers and Models - Lower level APIs to provide the most power and control

-> Pipelines are used for:
1. Sentiment analysis
2. Classifier
3. Named Entity Recognition
4. Question Answering
5. Summarizing
6. Translation
